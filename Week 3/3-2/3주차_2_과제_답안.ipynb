# 2. boxplot

2. 배기량_100단위 열은 기존의 배기량 데이터를 100으로 나눈 몫입니다.

# 3. folium

1. 창원시 와이파이 데이터가 제공됩니다.
2. 창원시 지도에 와이파이가 있는 곳을 나타내는 마커를 넣어보세요.


```python
import folium
창원광장 = [35.226367, 128.682232]

### 이 사이에 코드 입력
m = folium.Map(location=창원광장, zoom_start=12)

와이파이 = pd.read_excel("와이파이.xlsx")

위도리스트 = 와이파이["위도"]
경도리스트 = 와이파이["경도"]

for i in range(len(위도리스트)):
    folium.Marker(location=[위도리스트[i], 경도리스트[i]]).add_to(m)

###

m
```

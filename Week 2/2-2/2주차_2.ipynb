```python
# 참고 : requests 라이브러리 설명
# 쉽게 말해 '웹 브라우저'의 기능을 프로그래밍으로 구현할 수 있는 도구입니다.
# 웹사이트 접속, 로그인, 파일 다운로드가 가능합니다.

import requests

def 다운로드(파일명, 링크):
    with open(파일명, "wb") as f:
        res = requests.get(링크)
        f.write(res.content)

파일명 = "교통량.xlsx"
다운로드_링크 = "https://github.com/tyjeon24/keri-pythontraining/raw/main/Week%202/2-2/%EC%B0%BD%EC%9B%90%EC%8B%9C%20%EA%B5%90%ED%86%B5%EB%9F%89.xlsx"
다운로드(파일명, 다운로드_링크)
```

```python
# 1. 패키지 import
import pandas as pd # pandas.py 라는 파이썬 파일을 pd 라고 줄여쓰겠다는 의미입니다.
pd.options.display.float_format = '{:.2f}'.format # 소숫점 둘째자리까지 표현
```

```python
# 2. 데이터 읽기, 인덱스 설정
# 데이터를 읽고, 인덱스를 설정합니다.

교통량 = pd.read_excel("교통량.xlsx")
교통량 = 교통량.set_index("번호", drop=True)
교통량
```

```python
# 3. 데이터 요약 : describe()
# 데이터에 대한 요약을 제공합니다.

교통량.describe()
```

```python
# 4. 데이터 정렬
# 4.1. 특정 열 : sort_values(by=데이터)

교통량_정렬 = 교통량.sort_values(by="2021년 하행")
교통량_정렬
```

```python
# 4.2. 인덱스 기준 : sort_index()
교통량_정렬.sort_index()
```

```python
# 5. 열 이름 바꾸기 rename(columns={"기존이름":"바꿀이름"})

교통량 = 교통량.rename(columns={"관측지점":"장소"})
교통량
```

```python
# 6. 조건 인덱싱
# 6.1. 숫자 범위로 인덱싱 : 교통량[교통량["2021년 상행"] > 2000]

교통량[교통량["2021년 상행"] > 2000]
```

```python
# 6.2. 문자로 인덱싱 : str.contains("찾으려는 문자")
교통량[교통량["장소"].str.contains("북면")]
```

```python
# 7. 행, 열 인덱싱 나아가기 : 교통량.loc[1:3, ["A", "B"]]

교통량.loc[1, "장소"] # "장소"에 대괄호 포함하지 않음!
```
